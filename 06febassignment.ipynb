```python
1: A Python function that takes the given list as an argument and
returns the product of all the numbers..
   (including numeric keys and values of the dictionary) in a flat list 
    created from the input list...
    
    
def product_of_list(l):
    flat_list = []
    for item in l:
        if isinstance(item, (int, float)):
            flat_list.append(item)
        elif isinstance(item, (list, tuple)):
            flat_list += product_of_list(item)
        elif isinstance(item, dict):
            for key, value in item.items():
                if isinstance(key, (int, float)):
                    flat_list.append(key)
                if isinstance(value, (int, float)):
                    flat_list.append(value)
                elif isinstance(value, (list, tuple)):
                    flat_list += product_of_list(value)
    product = 1
    for num in flat_list:
        product *= num
    return product



>>> list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34),
             {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45,  22, 61, 34)}, 
             [56, 'data science'], 'Machine Learning']
>>> product_of_list(list1)
1207790080000



```

```python
2: A Python program that encrypts a message by replacing each letter with
the letter that precedes it in the alphabet, and replacing whitespaces with
a dollar sign...


def encrypt_message(message):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    encrypted_message = ''
    message = message.lower()
    for char in message:
        if char in alphabet:
            char_index = alphabet.index(char)
            encrypted_message += alphabet[char_index - 1]
        elif char == ' ':
            encrypted_message += '$'
        else:
            encrypted_message += char
    return encrypted_message



>>> message = "I want to become a Data Scientist."
Output:
>>> encrypt_message(message)
'h$vmt$sm$btnnf$a$czrr$sbhvr$sdtirt'





















```
